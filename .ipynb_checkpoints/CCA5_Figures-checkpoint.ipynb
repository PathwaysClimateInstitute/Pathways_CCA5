```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
from rasterio.enums import Resampling

# Path to your large raster
tif_path = r"C:\Users\MeaganBrown\Documents\ArcGIS\Projects\delete_later\PathwaysBayAreaGWMerged\GW_DTW_SLR_0.tif"

# Open and read a downsampled version
with rasterio.open(tif_path) as src:
    scale = 0.02  # Use 2% of full size – adjust if you want better/faster preview
    out_shape = (int(src.height * scale), int(src.width * scale))

    data = src.read(
        1,
        out_shape=out_shape,
        resampling=Resampling.average
    ).astype("float32")

    nodata = src.nodata
    if nodata is not None:
        data[data == nodata] = np.nan

# Reclassify for categories
reclassed = np.full(data.shape, np.nan)
reclassed[data < 0] = 0
reclassed[(data >= 0) & (data < 3)] = 1
reclassed[(data >= 3) & (data < 6)] = 2
reclassed[(data >= 6) & (data < 9)] = 3
reclassed[data >= 9] = 4

# Color and class mapping
colors = ['purple', '#FFDAB9', '#FFA500', '#FFDAB9', 'yellow']
labels = ['< 0 (Emergent)', '0–3 ft', '3–6 ft', '6–9 ft', '> 9 ft']
cmap = ListedColormap(colors)
norm = BoundaryNorm([-0.5, 0.5, 1.5, 2.5, 3.5, 4.5], cmap.N)

# Plot
plt.figure(figsize=(10, 8))
im = plt.imshow(reclassed, cmap=cmap, norm=norm)
cbar = plt.colorbar(im, ticks=[0, 1, 2, 3, 4])
cbar.ax.set_yticklabels(labels)
plt.title("Groundwater Depth Zones (Preview)")
plt.axis('off')
plt.tight_layout()
plt.show()



```
